Lets now mathematically detect outliers

Outlier detection can be done mathematically using two methods:-
- Standard Deviation Method (Z-score)
- Interquartile Range Method

**Here will use Interquartile Range Method as Z-score method is best suited when the data is uniformly distributed**

About IQR method:-

The IQR can be used to identify outliers by defining limits on the sample values that are a factor k of the IQR below the 25th percentile or above the 75th percentile. The common value for the factor k is the value 1.5. A factor k of 3 or more can be used to identify values that are extreme outliers or “far outs” when described in the context of box and whisker plots.

Lets now implement the above

**Outliers in Tenure**

```python
quartile_1, quartile_3 = np.percentile(customer_data['tenure'], [25, 75])
iqr = quartile_3 - quartile_1
lower_bound = (quartile_1 - (iqr * 1.5))
upper_bound = quartile_3 + (iqr * 1.5)

print("Lower bound for outliers:",lower_bound,"| Upper Bound for outliers",upper_bound)
print("No of outliers :" ,customer_data[((customer_data['tenure'] > upper_bound) | (customer_data['tenure'] < lower_bound))]['tenure'].count())
```

**Outliers in Monthly Charges**

```python
quartile_1, quartile_3 = np.percentile(customer_data['MonthlyCharges'], [25, 75])
iqr = quartile_3 - quartile_1
lower_bound = (quartile_1 - (iqr * 1.5))
upper_bound = quartile_3 + (iqr * 1.5)
print("Lower bound for outliers:",lower_bound,"| Upper Bound for outliers",upper_bound)
print("No of outliers :" ,customer_data[((customer_data['MonthlyCharges'] > upper_bound) | (customer_data['MonthlyCharges'] < lower_bound))]['MonthlyCharges'].count())
```

**Outliers in Total Charges**

```python
quartile_1, quartile_3 = np.percentile(customer_data['TotalCharges'], [25, 75])
iqr = quartile_3 - quartile_1
lower_bound = (quartile_1 - (iqr * 1.5))
upper_bound = quartile_3 + (iqr * 1.5)
print("Lower bound for outliers:",lower_bound,"| Upper Bound for outliers",upper_bound)
print("No of outliers :" ,customer_data[((customer_data['TotalCharges'] > upper_bound) | (customer_data['TotalCharges'] < lower_bound))]['TotalCharges'].count())
```

**Surprisingly , mathematically we could not find any outliers ,even though we didnt use a higher value of k.
It indicates that that the dataset doesnt have any extreme values**

**Continuous Variables (Multicollinearity)**

Detecting Multicollinearity using VIF (VIF determines the strength of the correlation between the independent variables)

About VIF:-

- VIF starts at 1 and has no upper limit

- VIF = 1, no correlation between the independent variable and the other variables

- VIF exceeding 5 or 10 indicates high multicollinearity between this independent variable and the others

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
calc_vif( customer_data[['tenure','TotalCharges','MonthlyCharges']])
```

**As expected the TotalCharges has high VIF,since it can be approximated as a function of other two variables,so we can exclude it from the model.**

**Categorical Variables (p-value)**

We can use that in conjunction with a Chi-squared test to make a more informed decision.

Note: the Chi-squared test can only be used to assess categorical variables.

We will need to understand p-values to interpret the Chi-squared output.

*The P-value indicates the % of more “extreme” observations if the null hypothesis is true and if we repeat the same experiment many times. The null hypothesis here is that the feature has no effect on the target (i.e. we should not use it as a predictor). Low p-values suggest we should reject our null hypothesis — in other words, reject the claim that the feature has no effect.*


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import SelectKBest, chi2


feature_selector = SelectKBest(chi2, k='all')
feature_selector.fit(X, y)
p_values = feature_selector.pvalues_
```

```python
p_above_thresh = [(i, p) for i, p in enumerate(p_values) if p > 0.05]
print(p_above_thresh)
```

The p-values for most of the features are very low. Only gender, MultipleLines, PhoneServicehave p-values > 0.05, which means that we do not have enough evidence to reject the claim that these features have no effect. This is well aligned with the story from the charts.

Note:- 

MultipleLines wont be dropped as for dummy variables ,all of them must have higher p-value.

For a better visual, let’s plot all categorical variables with p-value > 0.05.

```python
plt.bar(height=[i[1] for i in p_above_thresh],
x=X.columns[[i[0] for i in p_above_thresh]])
plt.xticks(rotation=45)
```

## Machine Learning Models

Note :-
**The Churn class is imbalanced as only 23% of the data is of the Churn customers.**

The challenge of working with imbalanced datasets is that most machine learning techniques will ignore, and in turn have poor performance on, the minority class, although typically it is performance on the minority class that is most important.

I have used the followin techinques Handling Imbalance:-

- Train test split with stratification
- SMOTE 

About SMOTE:-

Synthetic Minority Oversampling TEchnique, or SMOTE for short, works by selecting examples that are close in the feature space, drawing a line between the examples in the feature space and drawing a new sample at a point along that line.

Specifically, a random example from the minority class is first chosen. Then k of the nearest neighbors for that example are found (typically k=5). A randomly selected neighbor is chosen and a synthetic example is created at a randomly selected point between the two examples in feature space.

**1) Fitting the classifier object on the hyperparameters obtained using GridSearch on normal training data and training data after SMOTE**

**2) Print the details about the model classifaction report ,confusion matrix etc.**

**3) My personal insights about the model**

Note :-

1 ) The final peformace would be compared using **F1 score and AUC (the higher the better)** as its a good evaluation metric for classification models with imbalanced datasets

2 ) For all the models it was feasilble to show every step of hyperparameter tuning as some of the hyperparameters we selected after much experimenting

Lets split our dataset and count the no. of minority class samples in training and test data

```python
from collections import Counter


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(Counter(y_train))
print(Counter(y_test))
```

### Logistic Regression without SMOTE

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
result = lr.fit(X_train, y_train)
predictions = lr.predict(X_test)


from sklearn.metrics import classification_report,confusion_matrix,f1_score

print(classification_report(y_test,predictions))
f1_test = f1_score(y_test, predictions)
print('The f1 score for the testing data:', f1_test)
sns.heatmap(confusion_matrix(y_test,predictions),annot=True,fmt="d")
```

Tuning the hyperparameter classweight(for balancing imbalanced class)

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
lr = LogisticRegression()

#Setting the range for class weights
weights = np.linspace(0.0,0.99,200)

#Creating a dictionary grid for grid search
param_grid = {'class_weight': [{0:x, 1:1.0-x} for x in weights]}

#Fitting grid search to the train data with 5 folds
gridsearch = GridSearchCV(estimator= lr, 
                          param_grid= param_grid,
                          cv=StratifiedKFold(), 
                          n_jobs=-1, 
                          scoring='f1', 
                          verbose=2).fit(X_train, y_train)

#Ploting the score for different values of weight
sns.set_style('whitegrid')
plt.figure(figsize=(7,7))
weigh_data = pd.DataFrame({ 'score': gridsearch.cv_results_['mean_test_score'], 'weight': (1- weights)})
sns.lineplot(weigh_data['weight'], weigh_data['score'])
plt.xlabel('Weight for class 1')
plt.ylabel('f1')
plt.xticks([round(i/10,1) for i in range(0,11,1)])
plt.title('Scoring for different class weights', fontsize=24)
```

```python
gridsearch.best_params_
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(class_weight={0: 0.2885427135678392, 1: 0.7114572864321608})
result = lr.fit(X_train, y_train)
predictions = lr.predict(X_test)


from sklearn.metrics import classification_report,confusion_matrix,f1_score

print(classification_report(y_test,predictions))
f1_test = f1_score(y_test, predictions)
print('The f1 score for the testing data:', f1_test)
sns.heatmap(confusion_matrix(y_test,predictions),annot=True,fmt="d")
```

### Deep Neural Network with SMOTE

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Activation,Dropout
from tensorflow.keras.models import Model
from tensorflow import keras
```

```python
def create_model(learn_rate=0.01):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(25,activation='relu'))
    model.add(tf.keras.layers.Dropout(0.2))
    model.add(tf.keras.layers.Dense(15,activation='relu'))
    model.add(tf.keras.layers.Dense(15,activation='relu'))
    model.add(tf.keras.layers.Dropout(0.2))

    model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
    opt = keras.optimizers.Adam(learning_rate=learn_rate)
    model.compile(loss='binary_crossentropy', optimizer=opt)
    return model
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
model = KerasClassifier(build_fn=create_model)
learn_rate = [0.001,0.01,0.1]
batch_size = [10,32]
epochs = [10, 50]
param_grid = dict(learn_rate=learn_rate,batch_size=batch_size,epochs=epochs)

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=5,verbose=10,scoring='f1')
grid_result = grid.fit(X_train, y_train)

```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
nn = create_model(0.01)
history  = nn.fit(X_train, y_train, batch_size=32, epochs=15, validation_data=(X_test,y_test))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
nn.summary()
```

### Model Evaluation

```python
from sklearn import metrics

plt.figure(figsize=(10,10))

models = [
{'label': 'Logistic Regression', 'model': lr,},
{'label': 'Logistic Regression_Smote', 'model': lr_smote,},    
{'label': 'XGBOOST',    'model': xgb},  
{'label': 'XGBOOST_Smote',  'model': xgb_smote,},
{'label': 'RandomForest',  'model': rf,},
{'label': 'RandomForest_Smote','model': rf_smote,},
{'label': 'SVC', 'model': svc,},
{'label': 'SVC_Smote','model': svc_smote,},
{'label': 'NN','model': nn,},
{'label': 'NN_smote','model': nn_smote,}
]

# Below for loop iterates through your models list
for m in models:
    model = m['model'] # select the model
    y_pred=model.predict(X_test) # predict the test data
    # Compute False postive rate, and True positive rate
    if m['label'] == 'NN' or m['label'] == 'NN_smote':
        y_prob = model.predict_proba(X_test)[:,0]
    else:
        y_prob = model.predict_proba(X_test)[:,1]
    
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_prob)
# Calculate Area under the curve to display on the plot
    auc = metrics.roc_auc_score(y_test,model.predict(X_test))
# Now, plot the computed values
    plt.plot(fpr, tpr, label='%s ROC (AUC = %0.2f)' % (m['label'], auc))
# Custom settings for the plot 
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('1-Specificity(False Positive Rate)')
plt.ylabel('Sensitivity(True Positive Rate)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()   # Display
```

Inference :-

- **From the ROC curve above the highest AUC is obtained while using Neural Networks(AUC =0.84) . So the best model is Neural_network with SMOTE as for the same AUC is has a better F1 score than the one without smote**
